```
import time

def simulate_integrated_system(patient_data_vector, patient_id):
    # المرحلة 1: معالجة بيانات الحافة واستدعاء نموذج الذكاء الاصطناعي
    start_time = time.time()
    
    # تحضير وتشكيل قراءات المستشعرات الحية
    reshaped_data = np.array(patient_data_vector).reshape(1, -1)
    scaled_data = scaler.transform(reshaped_data)
    
    # التنبؤ الطبي الفوري على الحافة
    prediction_prob = edge_clf.predict(scaled_data, verbose=0)[0][0]
    diagnosis_result = "Heart Disease Confirmed" if prediction_prob > 0.5 else "Healthy / Clear"
    
    inference_end = time.time()
    t_inference = (inference_end - start_time) * 1000 # تحويل إلى مللي ثانية
    
    # المرحلة 2: التوثيق الفوري في شبكة البلوكشين الآمنة
    block_info = med_chain.add_medical_record(patient_id=patient_id, diagnosis=diagnosis_result, accuracy=float(prediction_prob))
    
    total_end = time.time()
    t_total = (total_end - start_time) * 1000 # الوقت إجمالي بالمللي ثانية
    
    return t_inference, t_total, block_info.hash

# فحص كفاءة النظام الكلية على حالة مريض عشوائية
sample_patient = np.random.randn(13)
t_inf, t_sys, b_hash = simulate_integrated_system(sample_patient, "PATIENT-2026")

print(f"[+] End-to-End System Evaluation Complete:")
print(f" - AI Inference Time at Edge: {t_inf:.2f} ms")
print(f" - Total System Latency: {t_sys:.2f} ms")
print(f" - Secure Ledger Receipt Hash: {b_hash}")
```
