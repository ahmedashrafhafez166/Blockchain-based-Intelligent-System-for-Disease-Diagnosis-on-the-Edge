```python
import hashlib
import json
import time

class MedicalBlock:
    def __init__(self, index, timestamp, patient_id, diagnosis, model_accuracy, previous_hash):
        self.index = index
        self.timestamp = timestamp
        self.patient_id = patient_id
        self.diagnosis = diagnosis
        self.model_accuracy = model_accuracy
        self.previous_hash = previous_hash
        self.nonce = 0
        self.hash = self.calculate_hash()

    def calculate_hash(self):
        # دمج البيانات لحساب البصمة الرقمية الفريدة للكتلة SHA-256
        block_string = (str(self.index) + str(self.timestamp) + str(self.patient_id) + 
                        str(self.diagnosis) + str(self.model_accuracy) + 
                        str(self.previous_hash) + str(self.nonce))
        return hashlib.sha256(block_string.encode()).hexdigest()

class MedicalBlockchain:
    def __init__(self):
        self.chain = [self.create_genesis_block()]
        
    def create_genesis_block(self):
        return MedicalBlock(0, time.time(), "GENESIS_PATIENT", "NONE", 1.0, "0")
        
    def get_latest_block(self):
        return self.chain[-1]
        
    def add_medical_record(self, patient_id, diagnosis, accuracy):
        # تحاكي هذه الدالة عمل الـ Smart Contract في أتمتة وتأمين البيانات
        latest_block = self.get_latest_block()
        new_block = MedicalBlock(
            index=latest_block.index + 1,
            timestamp=time.time(),
            patient_id=patient_id,
            diagnosis=diagnosis,
            model_accuracy=accuracy,
            previous_hash=latest_block.hash
        )
        # خوارزمية توافقية خفيفة (PoA) تناسب سرعة معالجة أجهزة الحافة الطبية
        while new_block.hash[:2] != "00":
            new_block.nonce += 1
            new_block.hash = new_block.calculate_hash()
            
        self.chain.append(new_block)
        return new_block

# تشغيل وتجربة وظائف البيئة المشفرة
med_chain = MedicalBlockchain()
start_tx = time.time()
block1 = med_chain.add_medical_record(patient_id="PT-9041", diagnosis="Heart Disease Detected", accuracy=0.9812)
end_tx = time.time()

print(f"[+] Secured Medical Block Created Successfully:")
print(f" - Block Index: {block1.index}")
print(f" - Generated Hash: {block1.hash}")
print(f" - Previous Hash: {block1.previous_hash}")
print(f" - Consensus Time: {end_tx - start_tx:.4f} seconds")
```
