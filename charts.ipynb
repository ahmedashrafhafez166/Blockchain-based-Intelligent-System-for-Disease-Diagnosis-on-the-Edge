```python
import matplotlib.pyplot as plt
import numpy as np

# البيانات المحدثة لتشمل الدراسة الأعلى دقة
studies = [
    'Al-Shammari\net al. (2021)', 
    'Priyanka\net al. (2023)', 
    'Abbas\net al. (2024)', 
    'Izhar\net al. (2023)',  # المرجع الأعلى دقة المضاف
    'Proposed\nFramework'
]

accuracy = [92.30, 95.70, 97.20, 99.70, 98.12]
# تم تعيين زمن تأخير مرتفع نسبياً للمرجع الأعلى دقة (120ms) نظراً لاعتماده الكامل على السحاب
latency = [45.00, 60.00, 15.60, 120.00, 10.52]  

x = np.arange(len(studies))
width = 0.35

fig, ax1 = plt.subplots(figsize=(11, 6))

# 1. المحور الأول: رسم أعمدة الدقة (Accuracy) باللون الأزرق
color1 = '#1f77b4'
ax1.set_xlabel('Comparative Studies / References', fontsize=12, fontweight='bold', labelpad=12)
ax1.set_ylabel('Classification Accuracy (%)', color=color1, fontsize=12, fontweight='bold')
bars1 = ax1.bar(x - width/2, accuracy, width, color=color1, alpha=0.85, label='Accuracy (%)')
ax1.tick_params(axis='y', labelcolor=color1)
ax1.set_ylim(70, 105)  # ضبط المدى لإبراز الفروق الدقيقة

# إضافة قيم الدقة فوق الأعمدة
for bar in bars1:
    height = bar.get_height()
    ax1.annotate(f'{height}%',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=9, fontweight='bold', color=color1)

# 2. المحور الثاني: رسم أعمدة زمن التأخير (Latency) باللون الأحمر
ax2 = ax1.twinx()  
color2 = '#d62728'
ax2.set_ylabel('Total System Latency (ms)', color=color2, fontsize=12, fontweight='bold')
bars2 = ax2.bar(x + width/2, latency, width, color=color2, alpha=0.85, label='Latency (ms)')
ax2.tick_params(axis='y', labelcolor=color2)
ax2.set_ylim(0, 140)  # رفع المدى لاستيعاب تأخير النظام السحابي

# إضافة قيم زمن التأخير فوق الأعمدة
for bar in bars2:
    height = bar.get_height()
    ax2.annotate(f'{height} ms',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=9, fontweight='bold', color=color2)

# ضبط العناوين والمظهر العام للبحث (Q1)
plt.title('Comprehensive Performance Trade-off: Accuracy vs. End-to-End Latency', fontsize=14, fontweight='bold', pad=20)
ax1.set_xticks(x)
ax1.set_xticklabels(studies, fontsize=9, fontweight='bold')
ax1.grid(True, axis='y', linestyle='--', alpha=0.5)

fig.tight_layout()
plt.show()
```
